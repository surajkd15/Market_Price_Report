```python
def scatterplot(feature):
    plt.figure(figsize=(14,6))
    plt.scatter(x='Price Date',y='Modal Price (Rs./Quintal)',data=data[data['Market Name']==feature])
    plt.xlabel('Date')
    plt.ylabel('Modal Price(Rs./Quintal)')
    plt.title('Price pattern for {} in yr 2020'.format(feature))
    plt.show()
```

```python
scatterplot('Achnera')
scatterplot('Agra')
scatterplot('Fatehabad')
scatterplot('Fatehpur Sikri')
```
